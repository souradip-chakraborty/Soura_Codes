```python
#### global vars
NSMAP = {'ab':'http://www.abbyy.com/FineReader_xml/FineReader10-schema-v1.xml'}
```

```python
def read_xml(xml_file):
    
    tree = ET.parse(xml_file) 
    root = tree.getroot()
    pages_text = []
    
    for page in root.findall('.//ab:page', namespaces=NSMAP):

        paras = []
        temp = ""
        for block in page.findall('ab:block', namespaces=NSMAP):
            for line in block.findall('ab:text/ab:par/ab:line/ab:formatting', namespaces=NSMAP):
                temp = ''.join([i if ord(i) < 128 else ' ' for i in temp + " " + line.text.strip()])

            paras.append(temp)
            temp = ""

        pages_text.append(paras)
    
    return pages_text
```

```python
def find_relevant_para(paras, word):
    
    for para_i in range(len(paras)):
        
        if word in paras[para_i].lower():
            
            if para_i < (len(paras) - 1):
                return paras[para_i] + paras[para_i + 1]
            else:
                return paras[para_i]
```

```python
###### code for train data
count_short = 0
count_long = 0

for file in df["File_Name"].tolist():
    
    print(file)
    
    pages_text = read_xml("/Users/s0c02nj/Desktop/Brainwaves/ML/data/"+file+".xml")
    
    flag = 0
    for p_i in range(len(pages_text)):
        
        page = " ".join(pages_text[p_i]).lower()
        
        if "additional termination event" in page:
            
            flag = 1
            
            #### get the paragraph which contains 
            if ("short term" in page  or "short-term" in page) and "party a" in page:
                
                print(find_relevant_para(pages_text[p_i], "short"))
                print("*******************\n\n")
                count_short +=1
            
            
            if ("long term" in page  or "long-term" in page)and "party a" in page:
                print(find_relevant_para(pages_text[p_i], "long"))
                print("*******************\n\n")
                count_long +=1
            
    if not flag:
        pass
        #print(file)
    
    
```
