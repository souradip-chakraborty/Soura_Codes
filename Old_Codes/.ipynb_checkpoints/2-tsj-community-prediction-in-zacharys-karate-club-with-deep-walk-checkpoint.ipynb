```python
from networkx import read_edgelist
from pandas import read_csv
from src.settings import RAW_DATA_DIR
from os.path import join as path_join
from src.data.datasets import load_karate_club
zkc = load_karate_club()
```

```python
from src.features.deep_walk import DeepWalk
from sklearn.metrics import classification_report, accuracy_score
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC

deep_walk = DeepWalk(zkc.meta_data['graph'], 10, 80, 2, 10)
classifier = LinearSVC()
model = make_pipeline(deep_walk, classifier)

model.fit(zkc.X_train, zkc.y_train)
preds = model.predict(zkc.X_test)

print(classification_report(zkc.y_test, preds))
print(accuracy_score(zkc.y_test, preds))
```
